```python
#Some introduction about Regular Expressions

sent = "Hello, My name is Rashu and  I like to eat Rashu's Icecream"

regex_name = re.compile('Rashu') # re object

regex_name.findall(sent)
```

```python
# substituiting some patterns

regex_name.sub('RASHU Tyagi',sent)
```

```python
names = ["Rashu", "Vashu", "Advik"]

for i in names:
  print(regex_name.sub(i,sent))
```

```python
sent = "Hello, My name is Rashu and  I like to eat Rahul's Icecream"

# \w --> means find any letter
regex_name = re.compile(r'R\whul')  # it means it will search for R then any letter then h then u then l.
regex_name.findall(sent)
```

```python
# pattern to replace 
# \d --> means python will look for any digit here.
"\\n\\n1\\n00:05.080 --> 00:09.310\\n"

pattern_to_replace = r'\\n\\n\d\\n\d\d:\d\d.\d\d\d --> \d\d:\d\d.\d\d\d\\n'

regex_timestamps = re.compile(pattern_to_replace)
regex_timestamps.findall(text)
```

### You can see that it is only finding those which have single digit numbers in the starting like n0,n1,n2 but unable to find n10,n11 etc.

```python
# \d{1,3} means that we are searching for a number which can be 1 digit or 2 digit or 3 digits.

pattern_to_replace = r'\\n\\n\d{1,3}\\n\d\d:\d\d.\d\d\d --> \d\d:\d\d.\d\d\d\\n'

regex_timestamps = re.compile(pattern_to_replace)
regex_timestamps.findall(text)
```

```python
newtext = regex_timestamps.sub(' ',text)
```

## Now next formatting which we will be doing is that we will find the 4 lettered words and replace all the characters except the first letter with "*".

```python
len(newtext)
```

```python
newtext.split() # words will be separated based on the " " by default.
```

```python
words = newtext.split()
print(words)
print(len(words))
```

```python
for word in words:
  print(f"The length of {word} is {len(word)}")
```

```python
# printing all the 4 lettered words
print("These are all the 4 lettered words")
for word in words:
  if len(word) == 4:
    print(word)
```

```python
# we are going to get words like own. , it's , etc as the 4 lettered words.
```

```python
# lets deal with them

# printing all the 4 lettered words
print("These are all the 4 lettered words")
for word in words:
  if len(word) == 4 and not word[-1]=="." and "'" not in word:
    print(word)
```

```python
# replacing the 4 lettered words as mentioned before
index = range(len(words))
for idx,word in zip(index,words):
  if len(word) == 4 and not word[-1]=="." and "'" not in word:
    words[idx] = word[0] + '***'
words  
```

```python
# Now we have received the set of words formatted in the fashion as we wanted. Now lets join them to
# form a single string.

final_clean_text = ' '.join(words)
final_clean_text
```

```python
# saving the file and downloading onto the local system.
with open('final_clean_file.txt','w') as f:
  f.write(final_clean_text)

files.download('final_clean_file.txt')
```
